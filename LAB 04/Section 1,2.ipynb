```python
#creating 2d array
arr_2d = np.array([[1,2,3,4],[5,6,7,8]],dtype="int")
print(np.sum(arr_2d))
print(np.mean(arr_2d))
print(np.min(arr_2d))
print(np.max(arr_2d))
print(np.std(arr_2d))
print(np.var(arr_2d))
```
