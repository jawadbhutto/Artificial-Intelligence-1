```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('titanic_cleaned.csv')
data
```

```python
data.shape
```

```python
data.head()
```

```python
X=data.drop(['Survived'], axis=1)
Y=data['Survived']
X.shape
```

```python
Y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
X_scaled=Scaler.fit_transform(X)
X_scaled
```

```python
X=pd.DataFrame(X_scaled, columns=X.columns)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=56, stratify=Y)
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
```

```python
clf=KNN(n_neighbors=12)
clf.fit(X_train,Y_train)
```

```python
test_predict=clf.predict(X_test)
K=f1_score(test_predict,Y_test)
print('Test F1 Score: ',K)
```

```python
def Elbow(K):
    test_error=[]
    for i in K:
        clf=KNN(n_neighbors=i)
        clf.fit(X_train,Y_train)
        y_predict=clf.predict(X_test)
        f1=f1_score(y_predict,Y_test)
        error=1-f1
        test_error.append(error)
    return test_error
```

```python
k=range(6,20,2)
```

```python
test=Elbow(k)
```

```python
plt.plot(k,test)
plt.xlabel('K Neighbors')
plt.ylabel('Test error')
plt.title('Elbow curve for test')
```

```python
clf=KNN(n_neighbors=10)
clf.fit(X_train,Y_train)
test_predict=clf.predict(X_test)
k=f1_score(test_predict,Y_test)
print('Test F1 Score: ',k)
```

```python
Y_test
```

```python
from sklearn.preprocessing import StandardScaler
Scalar=StandardScaler()
X_train_scaled=Scalar.fit_transform(X_train)
X_test_scaled=Scalar.transform(X_test)
X_train_scaled
```

```python
X_test_scaled
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
test_predict=clf.predict(X_test_scaled)
k=f1_score(test_predict,Y_test, average='macro')
print('F1 Score is: ',k)
```

```python
def Elbow(K):
    test_error=[]
    for i in K:
        clf=KNN(n_neighbors=i)
        clf.fit(X_train_scaled,Y_train)
        y_predict=clf.predict(X_test_scaled)
        f1=f1_score(y_predict,Y_test,average='macro')
        error=1-f1
        test_error.append(error)
    return test_error
```
