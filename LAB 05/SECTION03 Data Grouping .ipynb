```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("winequality-white.csv", sep=';')
#df = pd.DataFrame(data)
data
da = pd.DataFrame(data)
```

```python
qualty =  da.groupby("quality")
```

```python
qualty
```

```python
for quality, alcohol in qualty:
    print("Quality: ", quality)
    print("Alcohol Mean: ",alcohol["alcohol"].mean())
    print("\n")
```

```python
phlevel = da.groupby("pH")
```

```python
phlevel
```

```python
for ph, data in phlevel:
    print("pH : ",ph)
    print("Quality Mean: ",data["quality"].mean(),"\n")
```
