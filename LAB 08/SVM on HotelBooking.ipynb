```python
import pandas as pd
```

```python
HotelBooking = pd.read_csv("hotel_bookings_clean.csv")
```

```python
HotelBooking
```

```python
X = HotelBooking.drop('is_canceled', axis=1)
y = HotelBooking.is_canceled
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=56)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,f1_score
```

```python
c = classification_report(y_test,y_pred)
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test , y_pred)
```

```python
print(c)
print('Accuracy: ',acc)
print("F1 Score: ",f1)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
rs_param_grid = {
    "kernel" : ["linear","poly","rbf","sigmoid"],
    "C"  : [0.1, 1, 10],
    "gamma" : [0.00001, 0.0001, 0.001, 0.01, 0.1]
}
```

```python
clf = SVC()
```

```python
clf_rs = RandomizedSearchCV(
    estimator= clf,
    param_distributions = rs_param_grid,
    cv = 3,
    n_iter = 5,
    verbose = 2,
    random_state = 123 )
```

```python
clf_rs.fit(X_train, y_train)
```

```python
clf_rs.best_params_
```

```python
clf_rs.best_score_
```

```python
# Model Fitting on Best Params
```

```python
best_model = SVC(kernel='linear', C=1, gamma=0.0001)
```

```python
best_model.fit(X_train,y_train)
```

```python
y_pred = best_model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy: ",acc)
```

```python
f1 = f1_score(y_test, y_pred)
```

```python
print("F1 Score: ",f1)
```
